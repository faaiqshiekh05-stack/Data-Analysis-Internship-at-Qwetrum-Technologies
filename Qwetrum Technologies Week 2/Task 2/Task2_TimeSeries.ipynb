```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("D:/internship as data analyst/Qwetrum Technologies/Week 2/Task 2/Weather.csv")

# Plot 1: Raw data
plt.figure(figsize=(12,6))
plt.plot(df['temp_mean(c)'])
plt.title("Temperature Mean Over Records")
plt.savefig("graph_1_raw.png", dpi=300)
plt.show()

# Rolling 7
df['Rolling_7'] = df['temp_mean(c)'].rolling(window=7).mean()

plt.figure(figsize=(12,6))
plt.plot(df['temp_mean(c)'])
plt.plot(df['Rolling_7'])
plt.legend(['Temp Mean', '7-Day Average'])
plt.title("7-Day Rolling Average")
plt.savefig("graph_2_rolling7.png", dpi=300)
plt.show()

# Rolling 30
df['Rolling_30'] = df['temp_mean(c)'].rolling(window=30).mean()

plt.figure(figsize=(12,6))
plt.plot(df['temp_mean(c)'])
plt.plot(df['Rolling_30'])
plt.legend(['Temp Mean', '30-Day Average'])
plt.title("30-Day Rolling Average")
plt.savefig("graph_3_rolling30.png", dpi=300)
plt.show()

# Anomaly plot
mean = df['temp_mean(c)'].mean()
std = df['temp_mean(c)'].std()

upper = mean + 2*std
lower = mean - 2*std

anomalies = df[(df['temp_mean(c)'] > upper) | (df['temp_mean(c)'] < lower)]

plt.figure(figsize=(12,6))
plt.plot(df['temp_mean(c)'])
plt.scatter(anomalies.index, anomalies['temp_mean(c)'])
plt.title("Anomaly Detection")
plt.savefig("graph_4_anomalies.png", dpi=300)
plt.show()
```
