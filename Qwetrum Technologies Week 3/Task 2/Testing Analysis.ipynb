```python
import pandas as pd
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Group': ['A', 'B'],
    'Users': [1000, 1000],
    'Conversions': [120, 150]
})
data
data['Conversion_Rate'] = data['Conversions'] / data['Users']
data

conversions = np.array([120, 150])
users = np.array([1000, 1000])

stat, p_value = proportions_ztest(conversions, users)

print("Z-statistic:", stat)
print("P-value:", p_value)

plt.bar(data['Group'], data['Conversion_Rate'])
plt.title("A/B Test Conversion Rate Comparison")
plt.ylabel("Conversion Rate")
plt.show()
```
